## Data Cleaning Session 11 Assignemts

```python
#Import packages
import numpy as np
import pandas as pd
```

```python
#Create Data Frame
df=pd.DataFrame({'From_To':['LoNDon_paris','MAdrid_miLAN','lonDON_StockhOlm','Budapest_PaRis','Brussels_londOn'],'FlightNumber':[10045,np.nan,10065,np.nan,10085],'RecentDelays':[[23,47],[],[24,43,87],[13],[67,32]],'Airline':['KLM(!)','<Air France(12)','(British Airways.)','12.Air France','"Swiss Air"']})
df
```

#### 1. Adding Missing FlightNumber and convert Float to Integer

```python
## Adding missing FlightNumber
df['FlightNumber']=([ df['FlightNumber'][idx-1]+10 if pd.isnull(df['FlightNumber'][idx])  else df['FlightNumber'][idx]  for idx in df['FlightNumber'].index])
df
```

```python
## Convert FLight Number to Integer
df['FlightNumber']=df['FlightNumber'].astype(int)
df
```

#### 2. Split From_To column to two fields and assign to temporary DF

```python
## Splitting to From_To column to From and To
destination=df['From_To'].str.split("_",  expand = True) 
## Assigning values to a temporary DataFrame
df_tmp=pd.DataFrame(destination[:],index=df.index)
## Renaming Columns
df_tmp.columns=['From','To']
df_tmp
```

#### 3. Capitalize  From and To data

```python
## Capitalize from and to Data
df_tmp['From']=df_tmp['From'].str.capitalize()
df_tmp['To']=df_tmp['To'].str.capitalize()
df_tmp
```

#### 4. Delete From_To column and attach temporary DF 

```python
## Deleting From_To Column
df.drop('From_To',axis=1, inplace=True)
df
```

```python
## Attaching Temporary DF
df=df.join(df_tmp,how='left')
df
```

#### 5. Split and replace RecentDelays 

```python
## Splitting RecentDelays List and creating DF
df1=pd.DataFrame(df['RecentDelays'].values.tolist(), index= df.index,columns=['delay_1','delay_2','delay_3'])
```

```python
# Joining the new DF to the original DF
df=df.join(df1,how='left')
```

```python
# Dropping RecentDelays Column
df.drop(labels='RecentDelays', axis=1)
```
